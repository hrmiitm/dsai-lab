# Vectors in python and numpy

- np.array()
- np.arange()
- numpy.ndarray

```python
# Python Vectors
x = [1,2,3]
y = [4,5,6]

z = [x[i]+y[i] for i in range(len(x))]
z
```

```python
# NumPy Arrays
x = np.array([1,2,3]) # numpy.ndarray
y = np.arange(4, 7) # [4,5,6]
z = x + y
z
```

```python
# Operations
x = np.arange(1,4) # [1,2,3]
y = np.arange(4,7) # [4,5,6]
print(x+y)
print(x-y)
print(x*y)  # Hadamard Product
print(10*x) # Scaling Vector
```

# List vs NumPy Arrays

```python
x_lists = list(range(1, 1_000_001))
x_numpy = np.arange(1,1_000_001)
```

```python
%timeit [2*i for i in x_lists]
```

```python
%timeit 2*x_numpy
```

# Complex Operations on Arrays

```python
x = np.arange(1,5)
y = np.array([1, np.e, np.e**2, np.e**3])
print(x**2)
print(np.log10(x))
print(np.log(y))
print(x+5)
print(5*x + 2)
```

```python
# Dot Product
np.dot(x, y), x@y, x.T @ y
```
